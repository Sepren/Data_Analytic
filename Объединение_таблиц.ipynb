# Merge, join and concatenate

Классическая статья из документации Pandas

```python
import pandas as pd
```

```python
raw_data = {
    'subject_id': ['1', '2', '3', '4', '5'],
    'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a.index = [0, 1, 2, 3, 4]
df_a
```

```python
raw_data = {
    'subject_id': ['4', '5', '6', '7', '8'],
    'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b.index = [2, 3, 4, 5, 6]
df_b
```

```python
raw_data = {
    'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
    'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_n = pd.DataFrame(raw_data, columns = ['subject_id', 'test_id'])
df_n
```

# Concat
как правило используется для объединения таблиц по вертикальной или по горизонтальной оси

```python
df_new = pd.concat([df_a, df_b], axis=0)
df_new
```

```python
df_new_ = pd.concat([df_a, df_b], axis=1)
df_new_
```

```python
# Concat, но только с помощью атрибута 'inner'
df_new_ = pd.concat([df_a, df_b], axis=1, join='inner')
df_new_
```

# Append

Частный случай метода Concat с параметрами (axis=0, join='outer')

```python
df_a.append(df_b)
```

Метод **Join** основан на объединении таблиц через индексы (способ объединения указывается с помощью параметра how = ['left', 'right', 'inner', 'outer'])

```python
# df_a.join(df_b, how = 'left')
df_a.join(df_b, rsuffix='_right_table', how = 'left')
```

**Merge** используется для объединения таблиц по любым колонкам с помощью методов left_on и right_on

```python
df_new
```

```python
# Стоит обратить внимание, что колонка subject_id не дублируется в новой таблице
pd.merge(df_new, df_n, on='subject_id')
```

```python
pd.merge(df_new, df_n, left_on='subject_id', right_on='subject_id')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='left')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='right')
```

```python
pd.merge(df_a, df_b, right_index=True, left_index=True)
```
